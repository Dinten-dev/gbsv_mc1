## Day 2 – Methodological Design

### Theoretical Foundation and Method Choice
*Focus: principled justification aligned with the use case*

> This investigation applies the principle of the **Nyquist–Shannon Sampling Theorem** to the digital acquisition of SMI intraday price signals within the context of financial market surveillance in Switzerland.

**Nyquist–Shannon Sampling Theorem:** A continuous-time signal x(t) that is band-limited to a maximum frequency f_max can be perfectly reconstructed from its discrete samples if and only if the sampling rate satisfies:

$$f_s \geq 2 \cdot f_{\max}$$

The quantity $f_{\text{Nyquist}} = f_s / 2$ is the Nyquist frequency — the highest frequency that can be faithfully represented at sampling rate $f_s$. The minimum required sampling rate $f_s = 2 \cdot f_{\max}$ is called the Nyquist rate.

**Key assumptions:**
1. The signal must be **band-limited** (i.e., no frequency content above f_max)
2. Sampling must be **uniform** in time
3. The signal must be of **infinite duration** (in practice: sufficiently long relative to the period of interest)

> This principle assumes band-limitation and is valid under uniform sampling of stationary or quasi-stationary signals. The method of systematic sampling rate variation is selected because it directly operationalizes the Nyquist criterion by comparing signal reconstructions above, at, and below the critical rate. If these assumptions are violated — particularly the band-limitation assumption, which financial signals may violate due to transient events — aliasing may occur, compromising volatility estimation accuracy in this application.

### Parameter Definition and Mathematical Specification
*Focus: explicit parameter selection, derivation, and unit consistency*

> The signal is characterized by a sampling interval $T_s = 60$ s, resulting in a sampling rate $f_s = 1/T_s = 1/60$ Hz, and a total observation duration of approximately 7 trading days (~2500 data points).

**Bandwidth determination:** The bandwidth of the SMI 1-minute signal is estimated from its power spectral density. Based on the known properties of intraday financial data, relevant frequency components extend from approximately $f_{\min} \approx 1/(8.5 \cdot 3600) \approx 3.3 \times 10^{-5}$ Hz (one full trading session cycle) up to $f_{\max} \approx 1/120$ Hz (2-minute oscillation period). Price fluctuations faster than ~2 minutes are dominated by microstructure noise and are not physically meaningful for volatility estimation.

**Derived quantities:**
- Estimated maximum relevant frequency: $f_{\max} \approx 0.00833$ Hz
- Required Nyquist rate: $f_s \geq 2 \times 0.00833 = 0.01667$ Hz (i.e., $T_s \leq 60$ s)
- At the original 1-minute sampling: $f_{\text{Nyquist}} = 0.00833$ Hz, exactly matching our $f_{\max}$ estimate

> The value of $f_{\max}$ is derived from the known trading microstructure of the SIX Swiss Exchange, where meaningful price movements occur on time scales of minutes to hours. In this domain, the Nyquist frequency corresponds to the boundary between informative market dynamics and noise.

### Experimental Design for Next Days
*Focus: structured parameter variation and theoretical prediction*

> The baseline configuration is defined as the original 1-minute sampling ($f_s = 1/60$ Hz, i.e. downsampling factor $D = 1$).

The following downsampling factors will be varied systematically:

| Factor D | Effective interval | Effective f_s (Hz) | Condition relative to Nyquist |
|----------|-------------------|-------------------|------------------------------|
| 1        | 1 min             | 0.01667           | Above Nyquist rate (baseline)|
| 2        | 2 min             | 0.00833           | At Nyquist rate              |
| 5        | 5 min             | 0.00333           | Below Nyquist rate           |
| 15       | 15 min            | 0.00111           | Well below Nyquist rate      |
| 60       | 1 hour            | 0.000278          | Severely undersampled        |

> The selected range from $D=1$ to $D=60$ is justified because it spans from the original resolution through the Nyquist boundary down to a severely undersampled regime, allowing systematic observation of aliasing onset and progression.

> It is theoretically expected that:
> - At $D = 1$ (above Nyquist): faithful representation of all relevant frequency components.
> - At $D = 2$ (at Nyquist): marginal fidelity; highest frequency components are captured but with no margin for error.
> - At $D \geq 5$ (below Nyquist): aliasing occurs — high-frequency components fold back into the spectrum and appear as spurious low-frequency content, distorting both the time-domain waveform and the frequency-domain representation.

> All configurations will be documented to ensure traceability.

### Methodological Limitations and Risk Factors
*Focus: assumptions, stability, and potential misinterpretation*

> This approach assumes that the SMI price signal is approximately band-limited and quasi-stationary over the observation window, which may be violated during earnings announcements, central bank decisions, or geopolitical events that introduce transient, broadband frequency content.  
> The method is expected to be reliable when the market operates under normal trading conditions with gradual price evolution, but may become unstable or misleading during flash crashes or market halts, where the signal behavior departs significantly from stationarity.  
> In this financial surveillance use case, the primary risk factors are non-stationarity and event-driven transients, potentially leading to aliasing artifacts that could be misinterpreted as genuine low-frequency market cycles.
